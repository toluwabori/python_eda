```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
import numpy as np
import plotly.express as px
```

```python
df_crunchbase  = pd.read_csv('unicorns.csv')
```

```python
df_crunchbase
```

```python
df_crunchbase.info()
```

```python
df_crunchbase = df_crunchbase.drop(columns=['Updated at','Crunchbase Url','Date Joined','Investors','Company Website'])
```

```python
df_crunchbase.dropna(inplace=True)
```

```python
df_crunchbase['Year Joined']= df_crunchbase['Year Joined'].astype(object)
```

```python
df_crunchbase.info()
```

```python
df_crunchbase
```

```python
df_crunchbase['Year Joined'].value_counts()
```

```python
df_crunchbase['Year Joined'].value_counts().plot(kind = 'barh')
plt.xlabel('counts of companies')
plt.ylabel('year joined');
```

```python
#the list will be summarized to the largest 10
df_crunchbase['City'].value_counts().nlargest(10)
```

```python
df_crunchbase['City'].value_counts().nlargest(10).plot(kind='barh')
plt.xlabel('counts of companies')
plt.ylabel('CITY');
```

```python
df_crunchbase.Industry.value_counts()
```

```python
df_crunchbase.Industry.value_counts().plot(kind='barh')
plt.xlabel('counts of companies')
plt.ylabel('INDUSTRY');
```

```python
df_crunchbase.groupby('Country').sum().nlargest(5,columns = 'Last Valuation (Billion $)').plot(kind = 'barh');
```

```python
df_crunchbase.groupby('Industry').mean().nlargest(5,columns='Last Valuation (Billion $)').plot(kind='line',rot = 90);
```
